## Tarefa #1

Implemente funções em Python para aproximar a integral definida
$$
I = \int_a^b f(x)\,dx,
$$
usando:
i) a regra dos trapézios repetida/composta;
ii) a regra de Simpson repetida/composta.

a) Considere como parâmetros de entrada o espaçamento $h$ entre os nós (pontos) que serão usados e uma lista com os valores da função nesses nós: $f_j = f(a+j\, h)$,  $j=0,\dots,n$.

b) Considere como parâmetros de entrada o intervalo $[a, b]$, a função $f$ e a quantidade de nós (pontos) que serão usados.

**Observação**: Você precisará implementar 4 funções.


```python
# Função que calcula a integral utilizando o método do trapézio repetido
def trapezio_repetida(h, f):
    n = len(f) - 1
    integral = f[0] + f[n]  # Soma o primeiro e o último valor da função

    for i in range(1, n):  # Percorre os valores intermediários
        integral += 2 * f[i]  # Multiplica por 2 os valores intermediários

    integral *= h / 2  # Multiplica pelo passo da integração dividido por 2
    return integral

# Função que calcula a integral utilizando o método de Simpson repetido
def simpson_repetida(h, f):
    n = len(f) - 1
    integral = f[0] + f[n]  # Soma o primeiro e o último valor da função

    for i in range(1, n):  # Percorre os valores intermediários
        if i % 2 == 0:  # Verifica se o índice é par
            integral += 2 * f[i]  # Multiplica por 2 os valores de índice par
        else:
            integral += 4 * f[i]  # Multiplica por 4 os valores de índice ímpar

    integral *= h / 3  # Multiplica pelo passo da integração dividido por 3
    return integral

# Função que calcula a integral utilizando o método do trapézio composta
def trapezio_composta(a, b, f, n):
    h = (b - a) / n  # Calcula o tamanho do passo da integração
    x = [a + i * h for i in range(n + 1)]  # Cria uma lista de pontos equidistantes
    y = [f(xi) for xi in x]  # Calcula os valores da função nos pontos
    integral = trapezio_repetida(h, y)  # Chama a função trapezio_repetida com os valores calculados
    return integral

# Função que calcula a integral utilizando o método de Simpson composta
def simpson_composta(a, b, f, n):
    h = (b - a) / n  # Calcula o tamanho do passo da integração
    x = [a + i * h for i in range(n + 1)]  # Cria uma lista de pontos equidistantes
    y = [f(xi) for xi in x]  # Calcula os valores da função nos pontos
    integral = simpson_repetida(h, y)  # Chama a função simpson_repetida com os valores calculados
    return integral
```

##Tarefa #2: Teste as funções implementadas

a) Procure uma função $f$ que não seja um polinômio e cuja integral definida pode ser calculada analiticamente (de forma exata) no intervalo escolhido.

b) Obtenha aproximações do valor da integral definida dessa função considerando $n=10$, $50$, $100$ e $200$ (note que a quantidade de nós em cada caso será igual a $n+1$). Para cada valor de $n$ obtenha aproximações usando suas implementações das regras dos trapézios e de Simpson.

c) Determine os erros absolutos associados com as aproximações obtidas no item b).

d) Comente e explique os resultados obtidos em b) e c).

```python
import math

f = lambda x: math.sin(x) # Função f(x) = sen(x)

print("Letra A\n")
print("A função escolhida foi f(x) = sen(x)")

# Letra B e C

n_values = [10, 50, 100, 200]
integal_exata = 2

# Listas para armazenar os resultados
aproximacoes = []
erros_absolutos = []

for n in n_values:
    # Intervalo [0, pi]
    a = 0
    b = math.pi

    h = (b - a) / n # Calculando h baseado em a, b e n

    # Criando a lista de nós
    lista = []
    for i in range(n + 1):
        lista.append(f(a + i * h))

    # Aproximando os valores da integral
    aprox_rep_trapezio = trapezio_repetida(h, lista)
    aprox_rep_simpson = simpson_repetida(h, lista)
    aprox_comp_trapezio = trapezio_composta(a, b, f, n)
    aprox_comp_simpson = simpson_composta(a, b, f, n)

    # Armazenando os resultados nas listas
    aproximacoes.append((aprox_rep_trapezio, aprox_rep_simpson, aprox_comp_trapezio, aprox_comp_simpson))

    # Calculando os erros absolutos
    erro_rep_trapezio = abs(aprox_rep_trapezio - integal_exata)
    erro_rep_simpson = abs(aprox_rep_simpson - integal_exata)
    erro_comp_trapezio = abs(aprox_comp_trapezio - integal_exata)
    erro_comp_simpson = abs(aprox_comp_simpson - integal_exata)

    # Armazenando os erros absolutos nas listas
    erros_absolutos.append((erro_rep_trapezio, erro_rep_simpson, erro_comp_trapezio, erro_comp_simpson))


# Imprimindo as aproximações para todos os valores de n
print("\nLetra B\n")
for i, n in enumerate(n_values):
    aprox = aproximacoes[i]
    print(f"For n = {n}:")
    print(f"Aproximação repetida trapezio: {aprox[0]:.5f}")
    print(f"Aproximação composta trapezio: {aprox[2]:.5f}")
    print(f"Aproximação repetida simpson: {aprox[1]:.5f}")
    print(f"Aproximação composta simpson: {aprox[3]:.5f}")
    print("-----------------------------------------")

# Imprimindo os erros absolutos para todos os valores de n
print("\nLetra C\n")
for i, n in enumerate(n_values):
    erros = erros_absolutos[i]
    print(f"For n = {n}:")
    print(f"Erro absoluto da aproximação repetida trapezio: {erros[0]:.5f}")
    print(f"Erro absoluto da aproximação composta trapezio: {erros[2]:.5f}")
    print(f"Erro absoluto da aproximação repetida simpson: {erros[1]:.5f}")
    print(f"Erro absoluto da aproximação composta simpson: {erros[3]:.5f}")
    print("---------------------------------------------------------")
```
